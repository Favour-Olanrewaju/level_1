```python

import numpy as np
import pandas as pd

import sys

bicycle_weather = pd.read_csv('C:\\Datasets\\BicycleWeather.csv')
np.set_printoptions(threshold=sys.maxsize)

bicycle_weather
```

```python
# 1. Extract the PRCP column from the pandas object

prcp = np.array(bicycle_weather['PRCP'])
print(f" The PRCP Column: '\n' {prcp}")

# or

# prcp = bicycle_weather['PRCP'].values
# prcp
```

```python
# 2. Using boolean functions/operators separate numbers from NAN
nan_rows = bicycle_weather[bicycle_weather.isna().any(axis=1)]

non_nan_rows = bicycle_weather.dropna()

print(f"Rows with NaN values:'\n' {nan_rows}  '\n'")

print(f"Rows without NaN values:'\n' {non_nan_rows} '\n'")
```

```python
# 3. Using boolean masking separate all zeros and non-zeros into different variables.

non_zero = prcp[prcp != 0]
print(f" Non-zeros: '\n'{non_zero}")
```

```python
zero = prcp[prcp == 0]
print(f"zeros: '\n'{zero}")
```

```python
# 4. Multiplying non-zeros by 100 and dividing by 250

mul = (non_zero * 100)/250
print(mul)
```
